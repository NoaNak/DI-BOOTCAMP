```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
arr = np.random.randint(0, 75, size=100)
# Using NumPy, create an array of length 100 filled with random ints between 0 and 75
```

```python
arr
```

```python
arr_reshape = arr.reshape((50, 2))
# Reshape that array to a 50 x 2 matrix
```

```python
arr_reshape
```

```python
# Make a scatter plot of the 1st column as the x axis and the second column as the y axis
plt.scatter(arr_reshape[:, 0], arr_reshape[:, 1])
plt.scatter(arr_reshape[:, 0], arr_reshape[:, 1])
plt.title("Scatter Plot of Random Ints")
plt.xlabel("First Column")
plt.ylabel("Second Column")
```

```python
plt.hist(arr_reshape[:, 0])
plt.title("Histogram of First Column")
plt.xlabel("Value")
plt.ylabel("Frequency")
# Now make a histogram of each of the columns and make titles for each plot
```

```python
plt.hist(arr_reshape[:, 1])
plt.title("Histogram of Second Column")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```
